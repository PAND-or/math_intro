```python
def pol_to_dec(r, a):
    return (r * np.cos((np.pi * a) / 180), r * math.sin((np.pi * a) / 180))
 
pol_to_dec(1, 90)
```

```python
def dec_to_pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = (np.arctan2(y, x) * 180) / np.pi
    return(rho, phi)
    
dec_to_pol(0, 1)
```

## 1. Задание  
Напишите код, моделирующий выпадение поля в рулетке (с учетом поля зеро).

```python
def roulete():
    np.random.seed(21)
    roulet_list = list(range(0, 37)) 
    np.random.shuffle(roulet_list)                              
    rnd = random.randint(0, 36)
    roulet_dict = {}
    for k, i in enumerate(roulet_list):
        if i != 0:
            roulet_dict[i] = {
                'number': i,
                'color': 'red' if k % 2 == 0 else 'black'
            }
        else:
            roulet_dict[i] = {
                'number': 'zero',
                'color': ''
            }    
    user_chose = roulet_dict[rnd]
    return f'{user_chose["number"]} {user_chose["color"]}'

print(roulete())
```

```python
def sum_p(a, b):
    return a + b - (a * b)

print(f'Вероятность что при подбрасывании двух монет хоть у одной будет орел {sum_p(0.5, 0.5)}')
```
