# 1. Задание
Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
def vec_length(x):
    s = 0
    for i in x:
        s += i ** 2
    return math.sqrt(s)
```

```python
vec_length([20,20,20])
```

```python
vec_length([0,0,-20])
```

```python
def vec_sum(a, b):
    if len(a) != len(b):
        return 'Error a != b'
    
    res = []
    for i in range(len(a)):
        res.append(a[i] + b[i])
    return res
```

```python
vec_sum([20,20,20], [0,0,-20])
```

## 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:  
окружности,  
эллипса,  
гиперболы.  

```python
def my_circle(r, x0=0, y0=0):
    x = []
    y = []
    for t in np.linspace(0, 360, 361):
        x.append(r * np.cos((np.pi * t) / 180) + x0)
        y.append(r * np.sin((np.pi * t) / 180) + y0)
    plt.plot(x, y)
    plt.plot(x0, y0, marker='o')
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.axis('equal')
    plt.show()

my_circle(10, -3, 2)
```

```python
def my_elipse(a, b, x0=0, y0=0):
    x = []
    y = []
    for t in np.linspace(0, 360, 361):
        x.append(a * math.cos((np.pi * t) / 180) + x0)
        y.append(b * math.sin((np.pi * t) / 180) + y0)
    plt.plot(x, y)
    plt.plot(x0, y0, marker='o')
    plt.axis('equal')
    plt.show()

my_elipse(5, 3, -2, 2)
```

```python
def my_gip(a, b, x0=0, y0=0):
    x1 = []
    x2 = []
    y = []
    for t in np.linspace(-2, 2, 100):
        x1.append(a * math.cosh(t) + x0)
        x2.append(-a * math.cosh(t) + x0)
        y.append(b * math.sinh(t) + y0)
    plt.plot(x1, y)
    plt.plot(x2, y)
    plt.plot(x0, y0, marker='o')
    plt.show()

my_gip(3, 5, -5, 5)
```

```python
def dec_to_pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = (np.arctan2(y, x) * 180) / np.pi
    return(rho, phi)
    
dec_to_pol(0, 1)
```

### Решите систему уравнений и неравенств:  
y = x2 – 1  
exp(x) + x∙(1 – y) > 1  


```python
from scipy.optimize import fsolve
# exp(x) + x∙(1 – y) = 1
# exp(x) + x - x*y - 1 = 0
# exp(x) + x - 1 = x*y
# y = (exp(x) + x - 1) / x

x = np.linspace(-2,3,201)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

y = x2 – 1  
exp(x) + x∙(1 – y) = 1

```python
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y -1)

x1, y1 = fsolve(equations, (-1,1))
x2, y2 = fsolve(equations, (2,6))

print(x1, y1)
print(x2, y2)
```
